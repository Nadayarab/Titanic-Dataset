#Task 2: Reading and Preparing the Data

```python
import csv

data = []

with open(file_name, "r") as file:
    reader = csv.reader(file)
    for row in reader:
        data.append(row)

number_of_records = len(data) - 1
print("Number of records:", number_of_records)
```

#Task 3: Data Types and Type Conversion

```python
# -------------------------------
# Task 3: Data Types and Type Conversion
# -------------------------------

# دالة لتحويل القيم العددية مع التعامل مع القيم الفارغة أو الخاطئة
def convert_to_numeric(value, dtype=float):
    """
    Convert a string value to a numeric type (int or float).
    If the value is empty or invalid, return None.
    """
    try:
        return dtype(value)
    except ValueError:
        return None

# ==============================
# 1️⃣ تحديد نوع كل عمود (Text / Numeric)
# ==============================
headers = data[0]  # الهيدر
column_types = []

for col_index in range(len(headers)):
    sample_value = data[1][col_index]  # ناخذ أول قيمة في العمود
    try:
        float(sample_value)
        column_types.append("Numeric")
    except ValueError:
        column_types.append("Text")

print("Column Types:")
for i in range(len(headers)):
    print(f"{headers[i]:15} -> {column_types[i]}")
print("-" * 40)

# ==============================
# 2️⃣ تحويل الأعمدة الرقمية المهمة (Age و Fare)
# ==============================
ages = []
fares = []

for row in data[1:]:
    # تحويل Age
    age = convert_to_numeric(row[5], float)  # عمود 5 = Age
    ages.append(age)

    # تحويل Fare
    fare = convert_to_numeric(row[9], float)  # عمود 9 = Fare
    fares.append(fare)

# ==============================
# 3️⃣ التعامل مع القيم الفارغة: استبدالها بالمتوسط
# ==============================
valid_ages = [a for a in ages if a is not None]
average_age = sum(valid_ages) / len(valid_ages)
ages = [a if a is not None else average_age for a in ages]

valid_fares = [f for f in fares if f is not None]
average_fare = sum(valid_fares) / len(valid_fares)
fares = [f if f is not None else average_fare for f in fares]

# ==============================
# 4️⃣ فحص الأنواع بعد التحويل
# ==============================
print("Type check after conversion:")
print("Type of first Age value:", type(ages[0]))
print("Type of first Fare value:", type(fares[0]))

# ==============================
# 5️⃣ عرض أول 5 قيم بعد التحويل
# ==============================
print("First 5 Age values:", ages[:5])
print("First 5 Fare values:", fares[:5])
```

#Task 4: Calculations and Operations


```python
# -------------------------------
# Task 4: Calculations and Operations
# -------------------------------

# 1️⃣ Total number of passengers
total_passengers = len(data) - 1  # تجاهل الهيدر
print("Total number of passengers:", total_passengers)

# 2️⃣ Average Age (بعد التعامل مع القيم الفارغة من Task 3)
average_age = sum(ages) / len(ages)
print("Average age of passengers:", round(average_age, 2))

# 3️⃣ Highest and Lowest Fare
highest_fare = max(fares)
lowest_fare = min(fares)
print("Highest fare:", highest_fare)
print("Lowest fare:", lowest_fare)

# 4️⃣ استخدام comparison/logical operators
# مثال: عدد الركاب الذين أعمارهم أكبر من 30 و Fare أكبر من 50
passengers_over_30_high_fare = 0

for i in range(total_passengers):
    if ages[i] > 30 and fares[i] > 50:  # شرط منطقي AND
        passengers_over_30_high_fare += 1

print("Passengers older than 30 with fare > 50:", passengers_over_30_high_fare)

# 5️⃣ توضيح ترتيب العمليات باستخدام الأقواس
# مثال حساب معادلة وهمية: (Age * Fare) + (Fare / 2)
example_calc = (ages[0] * fares[0]) + (fares[0] / 2)
print("Example calculation (Age[0]*Fare[0] + Fare[0]/2):", round(example_calc, 2))

# مثال آخر يوضح تغيير ترتيب العمليات
example_calc_diff = ages[0] * (fares[0] + (fares[0] / 2))
print("Example with different parentheses:", round(example_calc_diff, 2))
```

#Task 5: User Input and Conditional Logic

```python
# -------------------------------
# Task 5: User Input and Conditional Logic
# -------------------------------

# طلب إدخال المستخدم
user_choice = input("Enter filter type (Pclass or Sex): ").strip()

# التحقق من صحة الإدخال
if user_choice.lower() not in ["pclass", "sex"]:
    print("Invalid input! Please enter 'Pclass' or 'Sex'.")
else:
    # طلب القيمة بناءً على نوع الفلتر
    if user_choice.lower() == "pclass":
        value = input("Enter passenger class (1, 2, or 3): ").strip()
        if value not in ["1", "2", "3"]:
            print("Invalid class! Must be 1, 2, or 3.")
        else:
            # تصفية الركاب حسب Pclass
            filtered_passengers = [row for row in data[1:] if row[2] == value]
            print(f"Total passengers in class {value}: {len(filtered_passengers)}")

    elif user_choice.lower() == "sex":
        value = input("Enter gender (male or female): ").strip().lower()
        if value not in ["male", "female"]:
            print("Invalid gender! Must be 'male' or 'female'.")
        else:
            # تصفية الركاب حسب الجنس
            filtered_passengers = [row for row in data[1:] if row[4].lower() == value]
            print(f"Total {value} passengers: {len(filtered_passengers)}")

    # مثال على عرض أول 5 ركاب بعد التصفية
    if 'filtered_passengers' in locals():
        print("\nFirst 5 filtered passengers (PassengerId, Name, Sex, Pclass):")
        for row in filtered_passengers[:5]:
            print(f"ID: {row[0]}, Name: {row[3]}, Sex: {row[4]}, Class: {row[2]}")
```

#Task 7: Using Data Structures

```python
# -------------------------------
# Task 7: Using Data Structures
# -------------------------------

# 1️⃣ استخدام tuple لتخزين الأعمدة (Headers)
headers_tuple = tuple(data[0])
print("Headers (tuple):", headers_tuple)

# 2️⃣ استخدام set لإيجاد القيم الفريدة في عمود Embarked
# index 11 = Embarked
embarked_values = set(row[11] for row in data[1:])
print("Unique embarkation ports:", embarked_values)

# 3️⃣ استخدام dictionary لتخزين ملخص عدد الركاب حسب Passenger Class
# index 2 = Pclass
passengers_by_class = {}

for row in data[1:]:
    pclass = row[2]
    if pclass in passengers_by_class:
        passengers_by_class[pclass] += 1
    else:
        passengers_by_class[pclass] = 1

# 4️⃣ استعراض محتوى الـ dictionary
print("\nNumber of passengers by class:")
for pclass, count in passengers_by_class.items():
    print(f"Class {pclass}: {count} passengers")
```

#Task 8: Loops and Flow Control

```python
# -------------------------------
# Task 8: Loops and Flow Control
# -------------------------------

# 1️⃣ استخدام for loop لمعالجة البيانات
# مثال: حساب عدد الركاب لكل فئة (Pclass)
passengers_by_class = {}
for row in data[1:]:
    pclass = row[2]
    if pclass in passengers_by_class:
        passengers_by_class[pclass] += 1
    else:
        passengers_by_class[pclass] = 1

print("Passengers by class using for loop:")
for pclass, count in passengers_by_class.items():
    print(f"Class {pclass}: {count}")

# -------------------------------
# 2️⃣ استخدام while loop للتحقق من صحة إدخال المستخدم
# مثال: إدخال الجنس
user_gender = ""
while user_gender.lower() not in ["male", "female"]:
    user_gender = input("Enter gender (male/female): ")
    if user_gender.lower() not in ["male", "female"]:
        print("Invalid input, try again.")

# -------------------------------
# 3️⃣ استخدام nested loops
# مثال: حساب عدد الركاب لكل فئة لكل جنس
# نعمل dictionary لكل جنس يحتوي على dictionary للفئات
gender_class_count = {}
genders = ["male", "female"]

for gender in genders:
    class_count = {}
    for row in data[1:]:
        if row[4].lower() == gender:
            pclass = row[2]
            if pclass in class_count:
                class_count[pclass] += 1
            else:
                class_count[pclass] = 1
    gender_class_count[gender] = class_count

print("\nNumber of passengers by gender and class:")
for gender, classes in gender_class_count.items():
    print(f"{gender.capitalize()}:")
    for pclass, count in classes.items():
        print(f"  Class {pclass}: {count}")

# -------------------------------
# 4️⃣ استخدام break و continue
# مثال: العثور على أول راكب أنثى في الفئة 1
for row in data[1:]:
    if row[4].lower() != "female":
        continue  # نتخطى الذكور
    if row[2] != "1":
        continue  # نتخطى الفئات الأخرى
    print("\nFirst female passenger in Class 1 found:")
    print(f"ID: {row[0]}, Name: {row[3]}, Age: {row[5]}, Fare: {row[9]}")
    break  # نوقف البحث بعد أول نتيجة
```
